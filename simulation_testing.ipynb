```python
import numpy as np
import matplotlib.pyplot as plt

from simulation import brownian_motion
```

## Antithetic variates

```python
n_sample = 8
t_grid = np.linspace(0, 2, 1000)
init_val = 2
drift = -2
covariance = 2


trajs = brownian_motion(
    n_sample=n_sample,
    t_grid=t_grid,
    init_val=init_val,
    drift=drift,
    covariance=covariance,
    antithetic_variates=True
)
```

```python
fig, ax = plt.subplots(figsize=(10, 7))

ax.plot(t_grid, init_val + drift * t_grid, 'k')
colors = ['b', 'g', 'r', 'm']
for c, traj in zip(colors, trajs[:4]):
    ax.plot(t_grid, traj, c=c)
for c, traj in zip(colors, trajs[4:]):
    ax.plot(t_grid, traj, c=c)
ax.grid()
```
